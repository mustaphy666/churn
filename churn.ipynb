```python
training_data=training_data.dropna()
training_data.head()
```

```python
encode=['Gender','Subscription Type','Contract Length']
```

```python
for c in encode:
    dummy=pd.get_dummies(training_data[c],prefix=c)
    training_data=pd.concat([training_data,dummy],axis=1)
    del training_data[c]
  
```

```python
train=training_data.drop(columns=['Churn','CustomerID'])
target=training_data['Churn']
```

```python
for c in encode:
    dummy=pd.get_dummies(testing_data[c],prefix=c)
    testing_data=pd.concat([testing_data,dummy],axis=1)
    del testing_data[c]  
```

```python
traint=testing_data.drop(columns=['Churn','CustomerID'])
targett=testing_data['Churn']
```

```python
traint
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(traint,targett,test_size=0.3,random_state=42)
```

```python
y_train
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_train)
from sklearn.metrics import accuracy_score
accuracy_score(y_train,pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
modell=RandomForestClassifier(max_depth=10,n_estimators=40,n_jobs=-1,random_state=42)
modell.fit(x_train,y_train)
```

```python
modell.score(x_train,y_train)
```

```python
modell.score(x_test,y_test)
```

```python
pred=modell.predict(x_test)[2]
pred
```
